```python
import random

# Define o estado inicial do quebra-cabeças
initial_state = [[2, 3, 1], [5, 8, 7], [6, 4, 0]]

# Verifica se o estado atual corresponde ao estado objetivo
def is_goal(state):
    goal_state = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
    return state == goal_state

# Gera um movimento aleatório válido
def generate_move(state):
    moves = []
    row, col = find_empty_space(state)

    # Verifica os movimentos possíveis
    if col > 0:
        moves.append((row, col - 1))  # Mover para a esquerda
    if col < 2:
        moves.append((row, col + 1))  # Mover para a direita
    if row > 0:
        moves.append((row - 1, col))  # Mover para cima
    if row < 2:
        moves.append((row + 1, col))  # Mover para baixo

    return random.choice(moves)

# Encontra a posição do espaço vazio
def find_empty_space(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

# Aplica um movimento ao estado atual
def apply_move(state, move):
    row, col = find_empty_space(state)
    new_row, new_col = move
    state[row][col], state[new_row][new_col] = state[new_row][new_col], state[row][col]

# Algoritmo de busca aleatória para resolver o quebra-cabeças
def random_search(initial_state):
    state = initial_state
    moves = 0
    max_iterations = 100000

    while not is_goal(state) and moves < max_iterations:
        move = generate_move(state)
        apply_move(state, move)
        moves += 1
        print(f"Move {moves}:")
        print_puzzle(state)
        print()

    if is_goal(state):
        print("Solução encontrada em", moves, "movimentos!")
    else:
        print("Limite máximo de iterações atingido. Não foi possível encontrar uma solução.")

# Função para imprimir o estado atual do quebra-cabeças
def print_puzzle(state):
    for row in state:
        print(" ".join(map(str, row)))

# Executa a busca aleatória
random_search(initial_state)
```
